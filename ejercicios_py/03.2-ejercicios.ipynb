# Ejercicio 1

El siguiente programa imprime por pantalla las palabras más frecuentes que se encuentre en una cadena de caracteres, en este caso una canción.
Sin embargo, tiene varios errores, tanto de sintaxis, como de ejecución, como de funcionalidad.

```python
with open('despacito.txt', 'r', encoding='utf-8') as file:
    lyrics = file.read()
```

```python
def split_into_words(lyrics):
    """
    Split a string into lowercase words, removing all punctuation characters,
    returning the result.
    """
    result = []
    for word in lyrics.lower().split(' '):  # lower() convierte las palabras en minusculas
        word = word.strip(',.;()"¡!')  # strip() elimina del inicio y del final los caracteres que le pasemos
        result.append(word)
    return result
```

```python
def words_to_frequencies(lyrics):
    """
    Convert words into frequencies. Return a dictionary whose keys are the
    words with the frequency as the value
    """
    freqs = {}
    for word in lyrics:
        if word in freqs:
            freqs[word] += 1
        else:
            freqs[word] = 1
    return freqs
```

```python
def most_common_words(frequencies):
    """
    Return a tuple containing:
    * The number of occurrences of a word in the first tuple element
    * A list containing the words with that frequency
    """
    values = frequencies.values()
    if not values:
        return (0, [])
    best = max(values)

    words = []
    for word, score in frequencies.items():
        if score == best:
            words.append(word)
    return (best, words)
```

```python
def get_more_often_user_words(frequencies, threshold=10):
    """
    Return a list of the words that are used more often, above
    the *optional* threshold. If no threshold is passed, use 10.
    """

    result = []
    while True:
       score = most_common_words(frequencies)
       if score[0] < threshold:
           break
       for w in score[1]:
           del frequencies[w]
       result.append(score)

    return result
```

```python
words_clean = split_into_words(lyrics)

freqs = words_to_frequencies(words_clean)

print("Palabras más frecuentes con más de 10 apariciones.")
print(get_more_often_user_words(freqs.copy()))
print("\n")
print("Palabras más frecuentes con más de 5 apariciones.")
print(get_more_often_user_words(freqs.copy(), threshold=5))
```
